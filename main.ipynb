## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    result = 1
    for number in range(1, n + 1):
        result *= number
    return result
final_result = factorial(4)
print(final_result)
        
        
#your code here
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ü§î`

```python
def unique(lst_un):
    unique_lst = []
    for numbers in lst_un:
        if numbers not in unique_lst:
            unique_lst.append(numbers)
    return unique_lst

final_list = unique([1, 2, 2, 3, 4])
print(final_list)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... üßê`

```python
def mode_counter(arr):
    lst = {}
    for numbers in arr:
        if numbers not in lst:
            lst[numbers] = 1
        else:
            lst[numbers] += 1
    
    max_number = 0
    winner = None

    for  number, count in lst.items():
        if count > max_number:
            max_number = count
            winner = number
    return winner
  
    

final_lst = mode_counter([1, 2, 3, 3, 3, 3, 3])
print(final_lst)
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. üòâ`

```python
def st_dev(list_sd):
    if len(list_sd) < 2:
        return 0.0
    
    total_sum = 0
    count = 0
    for number in list_sd:
        total_sum += number
        count += 1
    mean = total_sum / count
    
    sum_squares = 0
    for number in list_sd:
        difference = number - mean 
        sum_squares += difference ** 2
    
    variance = sum_squares / (count - 1)
    return variance ** 0.5
    

        
    
resultado = st_dev([1, 2, 3])
print(resultado)
```

```python
# This will test your function 
test_stdev(st_dev)
```
